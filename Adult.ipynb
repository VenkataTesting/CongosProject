```python
adult_df.replace("?", np.nan, inplace = True)
adult_df.dropna(inplace = True)
adult_df.head()
```

```python
#1.	How many men and women (sex feature) are represented in this dataset?
adult_df['gender'].value_counts()
```

```python
#What is the average age (age feature) of women?
adult_df[adult_df['gender'] == 'Female'].mean()['age']
```

```python
#What is the percentage of German citizens (native-country feature)?
(adult_df[adult_df['native-country']=='German'].count()/adult_df['native-country'].value_counts().sum())['native-country']
```

```python
age_mean=adult_df[adult_df['income'] >= '50K'].mean()['age']
print("Mean of age for income greater than 50K",age_mean)

age_std=adult_df[adult_df['income'] >= '50K'].std()['age']
print("Standard Deviation of age for income greater than 50K",age_std)

age_mean1=adult_df[adult_df['income'] <= '50K'].mean()['age']
print("Mean of age for income greater than 50K",age_mean1)

age_std1=adult_df[adult_df['income'] <= '50K'].std()['age']
print("Standard Deviation of age for income greater than 50K",age_std1)
```

```python
# Is it true that people who earn more than 50K have at least a high school education? (education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters or Doctorate feature)
adult_df[adult_df['income'] == '>50K']['education'].value_counts()
```
